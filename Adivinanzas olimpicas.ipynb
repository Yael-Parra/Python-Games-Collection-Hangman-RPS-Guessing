Juego de preguntas y respuestas: es un juego en el que se hacen preguntas sobre una variedad de temas y los jugadores deben responder correctamente para avanzar. Las preguntas pueden ser de diferentes categorÃ­as, como historia, ciencias, entretenimiento o cultura general. El objetivo del juego es responder la mayor cantidad de preguntas correctamente para ganar el juego.
```python
#DICCIONARIOS

# Ponemos todo en un solo diccionario. Si juntamos los diccionarios en una sola lista, no podremos acceder por key-value, ya que cada pregunta y respuesta nos queda como un valor independiente. 
# Para el juego de adivinanzas, tendremos 3 categorias: cultura general, deportes y mÃºsica y para ellos con 5 preguntas cada uno, siempre respetando la temÃ¡tica de los juegos olÃ­mpicos.

preguntas_respuestas_trivia= { # Olimpiadas 2024
    "Â¿En quÃ© ciudad se celebraran los juegos olimpicos de verano 2024?": "paris",
    "Â¿QuÃ© deporte fue incluido por primera vez en las olimpiadas de paris 2024?": "breaking",
    "Â¿CuÃ¡ntos deportes hubo en total en los juegos olimpicos de paris 2024?": "32",
    "Â¿cuantos paises participaron en los ultimos juegos olimpicos de verano en tokio 2020?": "206",
    "Â¿que deporte de equipo regreso en paris 2024 tras su ultima aparicion en londres 2012?": "balonmano",
    
    # Cultura General
    "Â¿En quÃ© ciudad se celebraron los primeros Juegos OlÃ­mpicos modernos en 1896?": "atenas",
    "Â¿QuÃ© paÃ­s ha ganado mÃ¡s medallas en la historia de los Juegos OlÃ­mpicos?": "estados unidos",
    "Â¿En quÃ© deporte se introdujo por primera vez la competencia femenina en los Juegos OlÃ­mpicos?": "tenis",
    "Â¿CuÃ¡ntos anillos aparecen en el logo de los Juegos OlÃ­mpicos?": "5 ",
    "Â¿CuÃ¡l es la ciudad que ha sido sede de los Juegos OlÃ­mpicos en tres ocasiones?": "londres ",

    # MÃºsica
    "Â¿QuÃ© canciÃ³n fue adoptada como himno oficial de los Juegos OlÃ­mpicos de 1960 en Roma?": "The Olympic Hymn",   
    "Â¿QuÃ© famoso cantante britÃ¡nico interpretÃ³ 'Imagine' en la ceremonia de apertura de los Juegos OlÃ­mpicos de Londres 2012?": "Paul McCartney",
    "Â¿CuÃ¡l fue la canciÃ³n mÃ¡s popular durante los Juegos OlÃ­mpicos de Barcelona 1992, que luego se convirtiÃ³ en un sÃ­mbolo de los juegos?": "Barcelona",
    "Â¿QuÃ© compositor fue el encargado de crear la mÃºsica oficial para los Juegos OlÃ­mpicos de 2008 en PekÃ­n?": "Tan Dun",
    "Â¿QuÃ© famosa canciÃ³n de la banda Queen se ha asociado con los Juegos OlÃ­mpicos como un himno de superaciÃ³n?": "we are the champions"}

lista_preguntas_trivia= list(preguntas_respuestas_trivia.keys())

nombre_jugador = input("Introduce nombre de jugador: ")

# Instrucciones del juego y bienvenida
print(f"HEY, {nombre_jugador}! ğŸ‰ Â¡BIENVENIDO/A A LOS JUEGOS OLÃMPICOS DEL TRIVIAL! ğŸ…\n\n"
      "Â¿Acaso te crees todo un campeÃ³n del conocimiento? Bueno, veamos si tienes lo que se necesita para subir al podio. ğŸ˜‰\n"
      "PrepÃ¡rate para enfrentar preguntas dignas de un atleta de la mente. ğŸ’ª\n"
      "Si respondes bien, sumarÃ¡s 1 punto y avanzarÃ¡s en tu camino hacia la gloria olÃ­mpica. ğŸ¥‡\n"
      "Si fallasâ€¦ bueno, vamos a ver cuÃ¡ntas caÃ­das resistes antes de besar la lona. Solo tienes 3 intentos, asÃ­ que Ãºsalos sabiamente. ğŸ¤”\n\n"
      "Recuerda, no todos nacen para ser leyendas olÃ­mpicas. Pero quizÃ¡s...     solo quizÃ¡s...    tÃº sÃ­. ğŸ†\n\n"
      "Â¿Listo? Â¿Listo para enfrentar el desafÃ­o, hÃ©roe del trivial? Â¡A la lÃ­nea de partida! ğŸƒğŸ’¨\n")



#FunciÃ³n para elegir de manera aleatoria las preguntas

import random

def eleccion_de_preguntas_random():
    # Seleeccion de pregunta aleatoria. Utilizamos la lista de preguntas, que seran keys del diccionario
    seleccionar_pregunta = random.choice(lista_preguntas_trivia)

    # Se elimina la pregunta random que ha salido de la lista para que no vuelva a salir
    lista_preguntas_trivia.remove(seleccionar_pregunta)

    return seleccionar_pregunta



#Vamos a poner un tope  mÃ¡ximo de preguntas, para que el juego no dure mucho tiempo, sobretodo para la presentaciÃ³n, haremos un mÃ¡ximo de 8 preguntas.
limite_partidas_acertadas = 5
limite_partidas_perdidas = 3
partidas_acertadas = 0
partidas_perdidas = 0

# Agregamos un marcador para que jugador pueda var como va 
marcador = []

# Se inicia el ciclo While
while partidas_acertadas < limite_partidas_acertadas and partidas_perdidas < limite_partidas_perdidas:

        # seleccion de pregunta random. Llamamos la funcion
        pregunta = eleccion_de_preguntas_random()

        # Respuesta del usuario
        respuesta_usuario = input(f" Vamos a ver que tanto sabes. Escribe la respuesta a esta pregunta:\n {pregunta.upper()}. Intenta ser muy clara en tu respuesta").lower()

        if respuesta_usuario ==  preguntas_respuestas_trivia[pregunta].lower():
                print (f"\nCorrecto!!! La respuesta a la pregunta {pregunta} es en efecto: {preguntas_respuestas_trivia[pregunta].upper()}")
                marcador.append(1)
                partidas_acertadas +=1
                print(f"Puntaje:\nPartidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas}  -  Partidas perdidas {partidas_perdidas}/{limite_partidas_perdidas}")
                print("___________________________________________________")
        else:
                print(f"\nÃ‰sta era un poco difÃ­cil...\nLa respuesta a la pregunta: {pregunta} es: {preguntas_respuestas_trivia[pregunta].upper()}")
                marcador.append(0)
                partidas_perdidas +=1
                print(f"Puntaje:\nPartidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas}  -  Partidas perdidas {partidas_perdidas}/{limite_partidas_perdidas}")
                print("___________________________________________________")
 

    # Imprimir resultados finales y marcador

print(f"Partidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas} - Partidas jugadas: {len(marcador)}")

if partidas_acertadas == 5:
        print("\nÂ¡Â¡Â¡Felicidades!!! Medalla de ORO para ti !ğŸ¥‡")
elif 4 <= partidas_acertadas <= 5:
        print("\nMuy bien! Medalla de PLATA para ti! ğŸ¥ˆ ")
elif 2 <= partidas_acertadas <= 3 :
        print("\nMuy bien! Medalla de BRONCE para ti! ğŸ¥‰ ")
elif partidas_acertadas < 2:
        print("\nHey!! No te desanimes, ningÃºn deportista llegÃ³ a las olimpiadas sin antes haber entrenado duro! Seguro a la prÃ³xima te va mejor ğŸ‹ï¸â€â™€ï¸")
    
    
```
