# Juego de preguntas y respuestas
_____________________________

Juego de preguntas y respuestas: es un juego en el que se hacen preguntas sobre una variedad de temas y los jugadores deben responder correctamente para avanzar. Las preguntas pueden ser de diferentes categorías, como historia, ciencias, entretenimiento o cultura general. El objetivo del juego es responder la mayor cantidad de preguntas correctamente para ganar el juego. Este juego puede ser jugado de forma individual o en equipo y es una forma divertida de aprender y poner a prueba tus conocimientos.

```python
#DICCIONARIOS

# Ponemos todo en un solo diccionario. Si juntamos los diccionarios en una sola lista, no podremos acceder por key-value, ya que cada pregunta y respuesta nos queda como un valor independiente. 
# Para el juego de adivinanzas, tendremos 3 categorias: cultura general, deportes y música y para ellos con 5 preguntas cada uno, siempre respetando la temática de los juegos olímpicos.

preguntas_respuestas_trivia= { # Olimpiadas 2024
    "¿En qué ciudad se celebraran los juegos olimpicos de verano 2024?": "paris",
    "¿Qué deporte fue incluido por primera vez en las olimpiadas de paris 2024?": "breaking",
    "¿Cuántos deportes hubo en total en los juegos olimpicos de paris 2024?": "32",
    "¿cuantos paises participaron en los ultimos juegos olimpicos de verano en tokio 2020?": "206",
    "¿que deporte de equipo regreso en paris 2024 tras su ultima aparicion en londres 2012?": "balonmano",
    
    # Cultura General
    "¿En qué ciudad se celebraron los primeros Juegos Olímpicos modernos en 1896?": "atenas",
    "¿Qué país ha ganado más medallas en la historia de los Juegos Olímpicos?": "estados unidos",
    "¿En qué deporte se introdujo por primera vez la competencia femenina en los Juegos Olímpicos?": "tenis",
    "¿Cuántos anillos aparecen en el logo de los Juegos Olímpicos?": "5 ",
    "¿Cuál es la ciudad que ha sido sede de los Juegos Olímpicos en tres ocasiones?": "londres ",

    # Música
    "¿Qué canción fue adoptada como himno oficial de los Juegos Olímpicos de 1960 en Roma?": "The Olympic Hymn",   
    "¿Qué famoso cantante británico interpretó 'Imagine' en la ceremonia de apertura de los Juegos Olímpicos de Londres 2012?": "Paul McCartney",
    "¿Cuál fue la canción más popular durante los Juegos Olímpicos de Barcelona 1992, que luego se convirtió en un símbolo de los juegos?": "Barcelona",
    "¿Qué compositor fue el encargado de crear la música oficial para los Juegos Olímpicos de 2008 en Pekín?": "Tan Dun",
    "¿Qué famosa canción de la banda Queen se ha asociado con los Juegos Olímpicos como un himno de superación?": "we are the champions"}

lista_preguntas_trivia= list(preguntas_respuestas_trivia.keys())

nombre_jugador = input("Introduce nombre de jugador: ")

# Instrucciones del juego y bienvenida
print(f"HEY, {nombre_jugador}! 🎉 ¡BIENVENIDO/A A LOS JUEGOS OLÍMPICOS DEL TRIVIAL! 🏅\n\n"
      "¿Acaso te crees todo un campeón del conocimiento? Bueno, veamos si tienes lo que se necesita para subir al podio. 😉\n"
      "Prepárate para enfrentar preguntas dignas de un atleta de la mente. 💪\n"
      "Si respondes bien, sumarás 1 punto y avanzarás en tu camino hacia la gloria olímpica. 🥇\n"
      "Si fallas… bueno, vamos a ver cuántas caídas resistes antes de besar la lona. Solo tienes 3 intentos, así que úsalos sabiamente. 🤔\n\n"
      "Recuerda, no todos nacen para ser leyendas olímpicas. Pero quizás...     solo quizás...    tú sí. 🏆\n\n"
      "¿Listo? ¿Listo para enfrentar el desafío, héroe del trivial? ¡A la línea de partida! 🏃💨\n")



#Función para elegir de manera aleatoria las preguntas

import random

def eleccion_de_preguntas_random():
    # Seleeccion de pregunta aleatoria. Utilizamos la lista de preguntas, que seran keys del diccionario
    seleccionar_pregunta = random.choice(lista_preguntas_trivia)

    # Se elimina la pregunta random que ha salido de la lista para que no vuelva a salir
    lista_preguntas_trivia.remove(seleccionar_pregunta)

    return seleccionar_pregunta



#Vamos a poner un tope  máximo de preguntas, para que el juego no dure mucho tiempo, sobretodo para la presentación, haremos un máximo de 8 preguntas.
limite_partidas_acertadas = 5
limite_partidas_perdidas = 3
partidas_acertadas = 0
partidas_perdidas = 0

# Agregamos un marcador para que jugador pueda var como va 
marcador = []

# Se inicia el ciclo While
while partidas_acertadas < limite_partidas_acertadas and partidas_perdidas < limite_partidas_perdidas:

        # seleccion de pregunta random. Llamamos la funcion
        pregunta = eleccion_de_preguntas_random()

        # Respuesta del usuario
        respuesta_usuario = input(f" Vamos a ver que tanto sabes. Escribe la respuesta a esta pregunta:\n {pregunta.upper()}. Intenta ser muy clara en tu respuesta").lower()

        if respuesta_usuario ==  preguntas_respuestas_trivia[pregunta].lower():
                print (f"\nCorrecto!!! La respuesta a la pregunta {pregunta} es en efecto: {preguntas_respuestas_trivia[pregunta].upper()}")
                marcador.append(1)
                partidas_acertadas +=1
                print(f"Puntaje:\nPartidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas}  -  Partidas perdidas {partidas_perdidas}/{limite_partidas_perdidas}")
                print("___________________________________________________")
        else:
                print(f"\nÉsta era un poco difícil...\nLa respuesta a la pregunta: {pregunta} es: {preguntas_respuestas_trivia[pregunta].upper()}")
                marcador.append(0)
                partidas_perdidas +=1
                print(f"Puntaje:\nPartidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas}  -  Partidas perdidas {partidas_perdidas}/{limite_partidas_perdidas}")
                print("___________________________________________________")
 

    # Imprimir resultados finales y marcador

print(f"Partidas ganadas: {partidas_acertadas}/{limite_partidas_acertadas} - Partidas jugadas: {len(marcador)}")

if partidas_acertadas == 5:
        print("\n¡¡¡Felicidades!!! Medalla de ORO para ti !🥇")
elif 4 <= partidas_acertadas <= 5:
        print("\nMuy bien! Medalla de PLATA para ti! 🥈 ")
elif 2 <= partidas_acertadas <= 3 :
        print("\nMuy bien! Medalla de BRONCE para ti! 🥉 ")
elif partidas_acertadas < 2:
        print("\nHey!! No te desanimes, ningún deportista llegó a las olimpiadas sin antes haber entrenado duro! Seguro a la próxima te va mejor 🏋️‍♀️")
    
    
```
